```python
import numpy as np
import matplotlib.pyplot as plt

def graph(formula, x, formula2, x2):
    y = formula(x)
    y2 = formula2(x2)
    plt.plot(x, y)
    plt.plot(x2, y2)
    plt.show()

graph(lambda x: -np.log(1-x), np.arange(-10, .99, .001), lambda x: -np.log(x), np.arange(.01, 10, .001))
```
