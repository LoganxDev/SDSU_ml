```python
import numpy as np
import matplotlib.pyplot as plt

def graph(formula, x_dom):
	x = np.array(x_dom)
	y = formula(x)
	plt.plot(x, y)
	plt.show()

graph(lambda x: x*0.5 + 30, range(0, 20))

```

```python
import numpy as np
import matplotlib.pyplot as plt

def checkVal(x):
    if x > 0:
        return -np.log(x)
    else:
        return -np.log(1-x)

def graph(x_dom):
    x = np.array(x_dom)
    vec = np.vectorize(checkVal)
    y = vec(x)
    plt.plot(x, y)
    plt.show()

graph(range(-100, 100))
```
